```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_community.tools import DuckDuckGoSearchRun
from datetime import datetime

GROQ_API_KEY = "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"
llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0, api_key=GROQ_API_KEY)
search = DuckDuckGoSearchRun()

# === Tools ===
def calculator(expression: str) -> str:
    try:
        allowed = set("0123456789+-*/(). ")
        if not all(c in allowed for c in expression):
            return "Invalid"
        return f"Result: {eval(expression)}"
    except:
        return "Galat expression"

def web_search(query: str) -> str:
    try:
        return search.run(query)
    except:
        return "Search nahi ho saka."

def detect_tool(text: str):
    import re
    # Calculator
    if any(op in text for op in ["+", "-", "*", "/"]):
        expr = re.findall(r"[\d\+\-\*\/\(\)\. ]+", text)
        if expr:
            return "calc", expr[0].strip()
    # Search
    search_words = ["kya hai", "kaun", "kab", "kyun", "news", "search", "batao", "latest", "who", "what", "when"]
    if any(w in text.lower() for w in search_words):
        return "search", text
    return "llm", text

# === Memory ===
history = [SystemMessage(content="Tu ek smart assistant hai. Web search aur calculator results use karke jawab de. Pichli baatein yaad rakh.")]

print("✅ Web Search + Calculator Agent Ready!\n")

while True:
    user_input = input("You: ").strip()
    if not user_input: continue
    if user_input.lower() in ["exit", "quit", "bye"]:
        print("👋 Goodbye!")
        break

    tool_name, tool_input = detect_tool(user_input)

    if tool_name == "calc":
        result = calculator(tool_input)
        print(f"\n🧮 {result}\n")
        continue

    if tool_name == "search":
        print("🔍 Internet pe search ho raha hai...")
        search_result = web_search(user_input)
        message = f"Sawal: {user_input}\n\nWeb Search Result:\n{search_result}\n\nIs information se jawab do."
    else:
        message = user_input

    history.append(HumanMessage(content=message))

    try:
        response = llm.invoke(history)
        history.append(AIMessage(content=response.content))
        print(f"\n🤖 {response.content}\n")
    except Exception as e:
        history.pop()
        print(f"❌ Error: {e}\n")
```
