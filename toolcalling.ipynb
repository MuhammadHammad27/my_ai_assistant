```python
from langchain_groq import ChatGroq
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
from langgraph.prebuilt import create_react_agent
from datetime import datetime

GROQ_API_KEY = "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"

llm = ChatGroq(
    model="llama-3.3-70b-versatile",  # ← yahi use karo
    temperature=0,
    api_key=GROQ_API_KEY
)

# =====================================
# Tools
# =====================================

@tool
def get_current_time() -> str:
    """Returns the current date and time."""
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

@tool
def calculator(expression: str) -> str:
    """
    Evaluates a math expression and returns the result.
    Example input: '5 + 3' or '10 * 2' or '100 / 4'
    """
    try:
        return str(eval(expression))
    except:
        return "Invalid expression!"

@tool
def search_knowledge(query: str) -> str:
    """
    Get information about Hammad.
    Args:
        query: one of these exact words: name, city, goal, learning, study_hours
    Returns:
        Information about Hammad
    """
    knowledge = {
        "name": "Hammad",
        "city": "Lahore", 
        "goal": "AI Engineer",
        "learning": "LangChain, RAG, Agents",
        "study_hours": "6-8 hours daily",
    }
    # Flexible matching
    query = query.lower().strip()
    for key in knowledge:
        if key in query or query in key:
            return f"{key}: {knowledge[key]}"
    return "Information not found"

tools = [get_current_time, calculator, search_knowledge]

# =====================================
# Agent
# =====================================

agent = create_react_agent(llm, tools)

print("✅ Agent Ready!\n")

# =====================================
# Test
# =====================================

print("🤖 Agent Ready!\n")

while True:
    user_input = input("You: ").strip()
    
    if not user_input:
        continue
    
    if user_input.lower() in ["exit", "quit", "bye"]:
        print("👋 Goodbye!")
        break
    
    result = agent.invoke({
        "messages": [HumanMessage(content=user_input)]
    })
    
    print(f"Bot: {result['messages'][-1].content}\n")
```
