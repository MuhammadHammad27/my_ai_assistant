```python
import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


nlp = spacy.load("en_core_web_sm")
```

```python
text = "The cat sat on the mat. The dog lay on the rug. The cat chased the dog. The dog barked at the cat."
print("original text:"+text)
```

```python
token_nltk = word_tokenize(text)
print("tokenized text:"+str(token_nltk))
```

```python
stopword=set(stopwords.words("english"))
filtered_tokens=[word for word in token_nltk if word not in stopword]
print("filtered tokens:"+str(filtered_tokens))
```

```python
stemmer=PorterStemmer()
stem_word=[stemmer.stem(word) for word in filtered_tokens]
print("stemmed tokens:"+str(stem_word))
```

```python
lemmi=WordNetLemmatizer()
lemmatizer_word=[WordNetLemmatizer().lemmatize(word) for word in filtered_tokens]
print("lemmatized tokens:"+str(lemmatizer_word))
```

```python
pos=[(token_nltk.text,token_nltk.pos_) for token_nltk in nlp(text)]
print("part of speech:"+str(pos))
```

```python
import random
from nltk.util import ngrams #to create grams
from collections import defaultdict
```

```python
bigrams=list(ngrams(token_nltk,2))
print(bigrams)
```

```python
word_dict = defaultdict(list)
for w1, w2 in bigrams:
  word_dict[w1].append(w2)

print(word_dict)
```

```python
def predict_next_word(word):
    if word in word_dict:
        return random.choice(word_dict[word])
    else:
        "Not found in the dictionary"
        
        
print(predict_next_word(input("Enter a word: ")))       
```
