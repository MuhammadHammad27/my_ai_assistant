```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate

GROQ_API_KEY = "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"   # ← Apni key daalo

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7,
    api_key=GROQ_API_KEY
)

# ====================== 1. ROLE PROMPTING ======================
print("=== 1. ROLE PROMPTING ===\n")

role_prompt = SystemMessage(content="""
You are a world-class Python teacher who explains concepts like a friend.
Explain in easy Hindi-English mix. Give examples.
""")

response = llm.invoke([role_prompt, HumanMessage(content="What is list comprehension?")])
print(response.content)
print("-"*50)

# ====================== 2. FEW-SHOT PROMPTING ======================
print("\n=== 2. FEW-SHOT PROMPTING ===\n")

few_shot_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    ("human", "Translate English to Hindi:\nEnglish: I love programming\nHindi:"),
    ("ai", "Mujhe programming bahut pasand hai"),

    ("human", "Translate English to Hindi:\nEnglish: How are you?\nHindi:"),
    ("ai", "Tum kaise ho?"),

    ("human", "Translate English to Hindi:\nEnglish: {english_text}\nHindi:")
])

chain = few_shot_template | llm
result = chain.invoke({"english_text": "Good morning, have a great day!"})
print(result.content)

# ====================== 3. CHAIN OF THOUGHT (CoT) ======================
print("\n=== 3. CHAIN OF THOUGHT ===\n")

cot_prompt = """
Solve this step by step:

Question: A shopkeeper buys a pen for Rs. 40 and sells it for Rs. 55. What is his profit percentage?

Think step by step and then give final answer.
"""

response = llm.invoke([HumanMessage(content=cot_prompt)])
print(response.content)

# ====================== 4. STRUCTURED OUTPUT (JSON) ======================
print("\n=== 4. STRUCTURED OUTPUT ===\n")

json_prompt = ChatPromptTemplate.from_template("""
Extract information in JSON format.

Text: {text}

Return in this format:
{{
  "name": "...",
  "age": number,
  "city": "...",
  "skills": ["skill1", "skill2"]
}}
""")

chain = json_prompt | llm
result = chain.invoke({"text": "My name is Rahul, I am 24 years old. I live in Delhi and I know Python, LangChain and Docker."})
print(result.content)
```
